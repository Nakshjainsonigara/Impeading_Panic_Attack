```python
import numpy as np
import pandas as pd
from pandas.io.formats.style import subset_args
```

```python
df=pd.read_excel("../data/1 Screening for depression.xlsx")
```

```python
df.rename(columns={df.columns[0]:"id"},inplace=True)
```

```python
(df.isna().sum()/df.shape[0])*100
```

```python
df.dropna(axis=0,how="all",inplace=True)
```

```python
df.columns=df.columns.str.lower()
```

```python
df[["id","survey start date","survey end date",'1.group scores for depression scale 1 summed up\t']]
```

```python
df=df.rename(columns={'1.group scores for depression scale 1 summed up\t':'depression_sum_score'})
```

```python
subset_data=df.columns[~df.columns.isin(["id","survey start date","survey end date","survey completion date","survey skipped"])]
df=df.dropna(axis=0,subset=subset_data,how="all")
```

```python
# we have droped data where id is missing casue there are no columns where id is na and depression score is there
df[(df["id"].isna())&(df["depression_sum_score"].notnull())]
df=df.dropna(subset=["id"])
```

```python
df[df["depression_sum_score"].isna()]
```

```python
df=df[df["id"].str.contains("SYM")]
```

```python
df["depression_sum_score"].isna().sum()
#now we can take final avergae
```

```python
df=pd.pivot_table(df,index="id",values="depression_sum_score",aggfunc="mean")
```
